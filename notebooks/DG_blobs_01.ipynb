```python
"""
TODO:
 - they don't seem to specify number of rows
 - this really doesn't have the options we seem to require at all
 - in center_box related to what they talk about in the paper?
"""

import itertools

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Defaults:
# n_samples=100, n_features=2, centers=None, 
# cluster_std=1.0, center_box=(-10.0, 10.0), 
# shuffle=True, random_state=None

# They do 10 of each
N_EACH = 1  

opts_nmin = [100]
opts_K = [2, 4, 8]
opts_dim = [2]  #, 4, 8]   # no. of features
opts_ov = [1.5] #, 5]      # overlap
opts_den = [1]  #, 4]
opts_nl = [0]   #, 0.1]     # noisy things (TODO)


def gen_dataset(nmin, no_clusters, dim, overlap, density, nl):
    """Generates individual dataset"""
    
    return make_blobs(
        n_samples=100, 
        n_features=dim,              # 2, 
        centers=no_clusters,         # None, 
        cluster_std=1.0, 
        center_box=(-10.0, 10.0), 
        shuffle=True, 
        random_state=None
        )


configs = itertools.product(opts_nmin, opts_K,
                            opts_dim, opts_ov,
                            opts_den, opts_nl)

for config in list(configs):
    i = 0
    
    while (i < N_EACH):
    
        data, labels = gen_dataset(*config)

        print(config)

        plt.scatter(data[:, 0], data[:, 1], c=labels)
        plt.show()
        
        i += 1
```
