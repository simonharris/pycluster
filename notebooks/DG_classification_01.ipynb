```python
"""
Synthetic dataset generation

TODO:
 - they don't specify number of rows
 - why do we have to have more features than clusters?
 - even the 2 dimensional plots don't look "right". Is this really 
     a suitable function to use?
 - I have no idea about many of the parameters
 - nor can I see how to use the parameters from the paper
"""

import itertools

import matplotlib.pyplot as plt
from sklearn.datasets import make_classification



# They do 10 of each
N_EACH = 1  

opts_nmin = [100]
opts_K = [2, 4, 8]
opts_dim = [2]  #, 4, 8]   # no. of features
opts_ov = [1.5] #, 5]      # overlap
opts_den = [1]  #, 4]
opts_nl = [0]   #, 0.1]     # noisy things (TODO)


def gen_dataset(nmin, no_clusters, dim, overlap, density, nl):
    """Generates individual dataset"""

    return make_classification(
        n_samples=100,
        n_features=dim,            # 20,
        n_informative=dim,         # 2,
        n_redundant=0,             # 2,
        n_repeated=0,
        n_classes=no_clusters,     # 2,
        n_clusters_per_class=1,    # 2,
        weights=None,
        flip_y=0.01,
        class_sep=1.0,
        hypercube=True,
        shift=0.0,
        scale=1.0,
        shuffle=False,             # True,
        random_state=None
        )


configs = itertools.product(opts_nmin, opts_K,
                            opts_dim, opts_ov,
                            opts_den, opts_nl)

for config in list(configs):
    i = 0
    
    while (i < N_EACH):
    
        data, labels = gen_dataset(*config)

        print(config)

        plt.scatter(data[:, 0], data[:, 1], c=labels)
        plt.show()
        
        i += 1
```
