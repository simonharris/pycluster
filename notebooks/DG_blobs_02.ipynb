```python
"""
TODO:
 - cardinality
"""

import itertools

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Defaults:
# n_samples=100, n_features=2, centers=None, 
# cluster_std=1.0, center_box=(-10.0, 10.0), 
# shuffle=True, random_state=None

opts_k = [2, 5, 10, 20]
opts_feats = [2] #, 10, 50, 100, 1000]
opts_samps = [1000]
opts_stdev = [0.5, 1, 1.5]


def gen_dataset(no_clusters, no_feats, no_samps, stdev):
    """Generates individual dataset"""
    
    return make_blobs(
        n_samples=no_samps, 
        n_features=no_feats,         
        centers=no_clusters,
        cluster_std=stdev)


configs = itertools.product(opts_k, opts_feats, opts_samps, opts_stdev)

for config in list(configs):

    data, labels = gen_dataset(*config)

    print(config)

    plt.scatter(data[:, 0], data[:, 1], c=labels)
    plt.show()
        
```
